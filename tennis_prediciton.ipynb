```python
df['tourney_date']=pd.to_datetime(df['tourney_date'].astype(str), format='%Y%m%d')
df['tourney_date'].head()   
```

```python
df.rename(columns=lambda x: x.replace('winner', 'player_a'), inplace=True)
df.rename(columns=lambda x: x.replace('loser', 'player_b'), inplace=True)


df.columns         
#df['tourney_name']=='Brisbane'
```

```python
features = ['tourney_name', 'surface', 'draw_size', 'tourney_level','match_num',
       'tourney_date',  'player_a_id', 'player_a_hand', 'player_a_ht', 'player_a_ioc', 'player_a_age',
       'player_b_id',  'player_b_hand',  'player_b_ht', 'player_b_ioc', 'player_b_age',  'best_of', 'round',       
       'player_a_rank', 'player_a_rank_points', 'player_b_rank', 'player_b_rank_points','minutes']
```

```python
df2 = df[features].copy()
#df2 = df2[df2['player_a_id']==103819]

df2.head()
```

```python
#df2['players']= df2[['player_a_id','player_b_id']].values.tolist()
df2['players_id']= df2['player_a_id']*df2['player_b_id']+df2['player_a_id']+df2['player_b_id']
df2.loc[ df2['players_id']==10724086899]
```

```python
df2 = df2.sort_values(['tourney_date', 'match_num'],ascending = [True, True])
df2['games_between']=  df2.groupby('players_id').cumcount()
df2['ratio_between_player_a']=  df2.groupby(['players_id','player_a_id']).cumcount()
df2['ratio_between_player_a']=df2['ratio_between_player_a']/df2['games_between']
df2['ratio_between_player_b']=1-df2['ratio_between_player_a']
df2['ratio_between_player_a']=df2['ratio_between_player_a'].fillna(0)
df2['ratio_between_player_b']=df2['ratio_between_player_b'].fillna(0)

df2['last_win_player_a'] = df2.groupby('players_id')['player_a_id'].shift()==df2['player_a_id']
df2['last_win_player_b'] = df2.groupby('players_id')['player_a_id'].shift()==df2['player_b_id']
#df2=df2.loc[df2['players_id']==10862673419] 

df2.head()
```

```python
players1= df2[['player_a_id','tourney_date', 'match_num']]
players1=players1.rename(columns={"player_a_id": "player_id"})
players1['win']=1
players2= df2[['player_b_id','tourney_date', 'match_num']]
players2=players2.rename(columns={"player_b_id": "player_id"})
players2['win']=0
players=pd.concat([players1, players2],ignore_index=True)
players = players.sort_values(['tourney_date', 'match_num'],
              ascending = [True, True])
players.head()
```

```python
players['games']=  players.groupby('player_id').cumcount()
players['winners'] =players.groupby('player_id')['win'].cumsum()-1
players.loc[players['winners']<0,'winners']=0
players['win_ration']=players['winners']/players['games']
players.loc[(players['win_ration']).isnull(),'win_ration']=0
players['first_game']= 1
players.loc[players['games']>0,'first_game']=0

players[['player_id','tourney_date','match_num','games','winners','win_ration','first_game']]
```

```python
df2=df2.merge(players[['player_id','tourney_date','match_num','games','winners','win_ration','first_game']]\
               ,left_on=['tourney_date','match_num','player_a_id'], right_on=['tourney_date','match_num','player_id'],how='left')

df2=df2.drop(columns=['player_id'])
df2=df2.rename(columns={"games": "player_a_games",'winners':'player_a_winners','win_ration':'player_a_win_ration','first_game':'player_a_first_game'})
df2.head()
```

```python
df2=df2.merge(players[['player_id','tourney_date','match_num','games','winners','win_ration','first_game']]\
               ,left_on=['tourney_date','match_num','player_b_id'], right_on=['tourney_date','match_num','player_id'],how='left')

df2=df2.drop(columns=['player_id'])
df2=df2.rename(columns={"games": "player_b_games",'winners':'player_b_winners','win_ration':'player_b_win_ration','first_game':'player_b_first_game'})
df2.head()
```
